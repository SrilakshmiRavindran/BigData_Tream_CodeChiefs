```python
import matplotlib.pyplot as plt

# Function to read data from the output TXT file
def read_data_from_txt(file_path):
    sales_data = {}
    with open(file_path, 'r') as file:
        for line in file:
            # Split the line into columns
            columns = line.strip().split('\t')  # Assuming tab-separated values
            name_parts = columns[0].split()  # Split the name into parts
            first_name = name_parts[0]
            last_name = name_parts[-1]  # Last part is assumed to be last name
            full_name = ' '.join(name_parts[:-1])  # Join remaining parts as first name or middle name

            # Store data in dictionary
            sales_data[full_name] = int(columns[1])  # Convert the number of cars sold to an integer
    return sales_data

# Path to the output TXT file from Hadoop
file_path = 'C:/Users/S560194/bdp1.txt'

# Read data from the TXT file
sales_data = read_data_from_txt(file_path)

# Plotting the data
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
plt.bar(sales_data.keys(), sales_data.values(), color='skyblue')
plt.title('Number of Cars Sold by Salesperson')
plt.xlabel('Salesperson')
plt.ylabel('Number of Cars Sold')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent overlap of labels
plt.show()
```

```python
import matplotlib.pyplot as plt

def read_data_from_txt(file_path):
    sales_data = {}
    with open(file_path, 'r') as file:
        for line in file:
            # Split the line into columns
            columns = line.strip().split('\t')  # Assuming tab-separated values
            if len(columns) >= 2:  # Check if there are at least 2 columns
                # Extract first name and last name
                full_name_parts = columns[0].split()
                first_name = full_name_parts[0]
                last_name = full_name_parts[-1]
                if len(full_name_parts) > 2:
                    # If there are more than 2 parts, assume the middle name is present
                    middle_name = ' '.join(full_name_parts[1:-1])
                    full_name = f"{first_name} {middle_name} {last_name}"
                else:
                    full_name = f"{first_name} {last_name}"
                # Extract commission data from the second column
                commission = float(columns[1])
                # Store data in dictionary
                sales_data[full_name] = commission
            else:
                pass  # Skip lines with insufficient columns
    return sales_data

# Path to the input TXT file
file_path = 'C:/Users/S560194/bdp2_subset.txt'

# Read data from the TXT file
sales_data = read_data_from_txt(file_path)

if sales_data:  # Check if sales_data is not empty
    # Plotting the data
    plt.figure(figsize=(12, 8))  # Adjust figure size as needed
    plt.bar(sales_data.keys(), sales_data.values(), color='skyblue')
    plt.title('Total Commission Earned by Salesperson')
    plt.xlabel('Salesperson')
    plt.ylabel('Total Commission Earned ($)')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.grid(True)
    plt.tight_layout()  # Adjust layout to prevent overlap of labels
    
    # Save the graph as an image file (e.g., PNG)
    plt.savefig('commission2_graph.png')
    
    plt.show()
else:
    print("No valid data to plot. All salespersons have missing commission data.")
```
